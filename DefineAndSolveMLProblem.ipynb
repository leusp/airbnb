```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.impute import SimpleImputer
```

```python
features = [
    'room_type', 'accommodates', 'bathrooms', 'bedrooms', 'beds',
    'neighbourhood_group_cleansed', 'review_scores_rating', 
    'number_of_reviews', 'instant_bookable', 'price'
]
df = df[features]

# Check for missing values
print(df.isnull().sum())

df = df.dropna(subset=['price'])
```

```python
# missing numerical values with median
num_features = ['accommodates', 'bathrooms', 'bedrooms', 'beds', 'review_scores_rating', 'number_of_reviews']
df[num_features] = df[num_features].fillna(df[num_features].median())

#  'instant_bookable' to binary 
df['instant_bookable'] = df['instant_bookable'].astype(int)
```

```python
#one hot encoding
cat_features = ['room_type', 'neighbourhood_group_cleansed']
df = pd.get_dummies(df, columns=cat_features, drop_first=True)
```

```python
X = df.drop('price', axis=1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

#### Linear Regression Model


```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)

rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression Results:")
print(f"RMSE: {rmse_lr:.2f}")
print(f"MAE: {mae_lr:.2f}")
print(f"R^2: {r2_lr:.2f}")
```

#### Random Forest Model

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("\nRandom Forest Results:")
print(f"RMSE: {rmse_rf:.2f}")
print(f"MAE: {mae_rf:.2f}")
print(f"R^2: {r2_rf:.2f}")
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

best_rf = grid_search.best_estimator_
print("\nBest Random Forest Parameters:", grid_search.best_params_)

y_pred_rf_tuned = best_rf.predict(X_test)
rmse_rf_tuned = np.sqrt(mean_squared_error(y_test, y_pred_rf_tuned))
r2_rf_tuned = r2_score(y_test, y_pred_rf_tuned)

print("\nTuned Random Forest Results:")
print(f"RMSE: {rmse_rf_tuned:.2f}")
print(f"RÂ²: {r2_rf_tuned:.2f}")
```

```python
results = {
    'Model': ['Linear Regression', 'Random Forest (Default)', 'Random Forest (Tuned)'],
    'RMSE': [rmse_lr, rmse_rf, rmse_rf_tuned],
    'RÂ²': [r2_lr, r2_rf, r2_rf_tuned]
}

results_df = pd.DataFrame(results)
print("\nModel Comparison:")
print(results_df)
```

```python
# Feature Importance (Random Forest)
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': best_rf.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance)
plt.title("Random Forest Feature Importance")
plt.show()
```

#### Conclusions
After removing rows with missing price values, I imputed missing values in the numerical columns using the median. Binary columns were converted to numerical format, and categorical variables were one-hot encoded using a ColumnTransformer. The dataset was split into training and test sets using an 80/20 split.

I trained and evaluated two models: Linear Regression and Random Forest Regressor. Performance was evaluated using three key regression metrics on the test data:

* Mean Absolute Error (MAE): This measures the average absolute difference between predicted and actual prices. Lower values indicate better performance.

* Root Mean Squared Error (RMSE): This penalizes larger errors more heavily than MAE and is useful when large deviations are particularly undesirable.

* R^2 Score: This represents the proportion of variance in the target variable explained by the model. A value closer to 1 indicates a better fit.

The Linear Regression model achieved a Root Mean Squared Error (RMSE) of approximately 107.72$ and an R^2 score of 0.41, meaning it was able to explain about 41 percent
of the variance in Airbnb listing prices. The default Random Forest model produced nearly identical performance, with an RMSE of $107.79 and an RÂ² score of 0.41, suggesting it did not offer a significant improvement over Linear Regression without tuning. However, after hyperparameter optimization, the tuned Random Forest Regressor achieved the best results, lowering the RMSE to $100.86 and increasing the RÂ² score to 0.49. This indicates that the tuned model explained nearly half the variance in listing prices and provided more accurate predictions overall. While the improvement is moderate, it shows that model tuning can lead to measurable gains in performance for this regression task.


Overall, the Random Forest Regressor was the best-performing model in this analysis, showing improved accuracy and generalization over the linear baseline.
